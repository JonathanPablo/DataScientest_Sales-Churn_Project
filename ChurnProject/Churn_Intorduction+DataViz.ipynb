## Introduction

### Idea

The data contains informations about contracts of international health insurance products. Including if they got terminated by the customer or not. 

Our goal is to predict, if - or the probability that - a contract gets terminated by the customer.
Further we want to gain insights in 
1. Main (global) influencing factors that lead to termination of a contract.
2. Top 5% of contracts that are most likely to be terminated in the near future. As well as
3. Main individual factors that could lead to termination of specific contracts.  

To get a first idea we can look at some other articles and projects:
- __[Predicting Travel Booking Cancellations with ML](https://analytx4t.com/predicting-travel-booking-cancellations-with-machine-learning/)__: Quick overview about what customer churn means and necesarry steps to make ML predictions.
- __[Github: Churn-Prediction](https://github.com/Debasishsaha123/Churn-Prediction)__: Explanations and examples of data exploration in churn prediction of telecomunication contracts on github.
- __[Kaggle: Hotel Booking Cancellations Predictive Analysis](https://www.kaggle.com/code/gwen08/hotel-booking-cancellations-predictive-analysis)__: End-to-End example for predictive analysis of hotel booking cancellations on kaggle.

### Data collection

A global overview over the (already existing and planned) feature columns as well as a sample of the first 100 lines can be found in the excel file __[target+feature_ideas](https://bdae0-my.sharepoint.com/:x:/g/personal/jonathan_leipold_bdae-gruppe_de/Ea3v0IowZTdNlwOxyIpLuysB7UtgEqo9PdLIB3Ty75MWIg?e=IaMsCe)__.

The data contains (for now) mainly informations about:
1. Contract: product, application-, start-, end-, (and if terminated:) termination-Date, termination reason
2. Customer: birthday, gender and nationality of insured person, country of contract holder
3. Paid premiums and claimed expenses: premium amount, claimed amount, number of claims (each as total sum and sum over last year)
4. Claim processing: payout amount, payout duration (total, last year and ratio to claimed amount)

In addition there are possibilities to collect even more informations about:
- last premium / claim: last treatment location, last premium Amount, last premium Country
- last premium adjustment (pA) of product: time since last pA, (mean) change of last pA

To collect even more features it could be possible to even collect informations from further sources like:
- Mail Server: Amount of received mails from isnured person / contract holder (in total / in the last year)

#### Replace NaN

```python
#get endDate as Minimum of updateDate (for active contracts) & effEndDate
endDate = contracts[['policy_effEndDate','update_Date']].min(axis=1) 
#endDate.isna().sum() # no NA
#contracts.policy_startDate.isna().sum() #no NA
#endDate.min() #minimum endDate (>2017, because contracts only contain data with StartDate >= 2017 )
```

```python
#some tests..
(endDate - contracts.policy_startDate) / np.timedelta64(1, 'M')
np.round((endDate - contracts.policy_startDate) / np.timedelta64(1, 'M'),0)
(np.round((endDate - contracts.policy_startDate) / np.timedelta64(1, 'M'),0)+1)
(np.round((endDate - contracts.policy_startDate) / np.timedelta64(1, 'M'),0)+1).isna().sum() 
(np.round((endDate - contracts.policy_startDate) / np.timedelta64(1, 'M'),0)+1).max()
np.round((endDate - contracts.policy_startDate) / np.timedelta64(1, 'Y'),0)
```

```python
# create contracts['policyAge[Y]'], drop outlayers & get max_policyAge_Months
contracts['policyAge[Y]'] = np.round((endDate - contracts.policy_startDate) / np.timedelta64(1, 'Y'),0)#.astype(int)
contracts['policyAge[Y]'].value_counts().sort_index() #1 contract has much higher age as all others. Check this one.
contracts.loc[contracts['policyAge[Y]'] > 7] #reason: update_Date == '2099-12-21'
contracts.loc[contracts['update_Date'] > str(today)] #only 1 contract. drop it.
contracts = contracts.loc[contracts['update_Date'] <= str(today)]
max_policyAge_Months = (np.round((endDate - contracts.policy_startDate) / np.timedelta64(1, 'M'),0)+1).max() #get in Months
max_policyAge_Months
```

```python
# create contracts['insured_Age'] & get max_Age
contracts['insured_Age'] = np.round((contracts.update_Date - contracts.insured_birthDate) / np.timedelta64(1, 'Y'),0).astype(int)
max_Age = contracts['insured_Age'].max()
max_Age
```

```python
#create function 'plot_categorical_influence' to display all informations the same way for different columns as influence
def plot_categorical_influence(df=None, target_col='num_claims_cat', hue_col=None):
    """
    Generate two plots to visualize the influence of a hue column on a categorical target column.

    Parameters:
        df (pd.DataFrame, optional): The DataFrame containing the data. If not provided, a sample DataFrame will be used.
        target_col (str, optional): The name of the categorical target column. Default is 'num_claims_cat'.
        hue_col (str): The name of the column to use as the hue variable.

    Returns:
        None (displays plots).

    Example:
        # Load your DataFrame 'contracts' here
        plot_categorical_influence(df=contracts, target_col='num_claims_cat', hue_col='another_column')
    """

    if df is None:
        # Create a sample DataFrame for demonstration
        data = {
            'num_claims_cat': ['Low', 'Medium', 'High'] * 30,
            'hue_col': ['A', 'B', 'C'] * 30
        }
        df = pd.DataFrame(data)

    # Calculate normalized ratios
    ratios = (
        df.groupby([hue_col, target_col])
        .size()
        .unstack(target_col, fill_value=0)
        .apply(lambda row: row / row.sum(), axis=1)
    )

    # Create subplots
    fig, axs = plt.subplots(1, 2, figsize=(15, 5), width_ratios=[1, 3])

    # Left plot: distplot of hue variable
    sns.countplot(data=df, x=hue_col, ax=axs[0])
    axs[0].set_title(f'Distribution of {hue_col}')
    axs[0].set_xlabel(hue_col)
    axs[0].set_ylabel('Count')

    # Add percentages in the middle of the bars in the left subplot
    for container in axs[0].containers:
        heights = [p.get_height() for p in container]
        total = sum(heights)
        for p in container:
            axs[0].annotate(f'{p.get_height()/total:.0%}', (p.get_x() + p.get_width() / 2., p.get_height() / 2.), ha='center', va='center', fontsize=9)
    
    # Right plot: horizontal barplot of normalized ratios with percentages
    ratios.plot(kind='barh', stacked=True, ax=axs[1])
    axs[1].set_title(f'Influence of {hue_col} on {target_col}')
    axs[1].set_xlabel('Normalized Ratio')
    axs[1].set_ylabel('')
    axs[1].legend(title=target_col)
    
    # Add percentages above the bars in the right subplot using bar_label
    for container in axs[1].containers:
        axs[1].bar_label(container, label_type='center', padding=2)
        
    # Display value count DataFrames for each hue value
    unique_hue_values = df[hue_col].unique()
    for hue_value in unique_hue_values:
        filtered_df = df[df[hue_col] == hue_value]
        value_counts_df = filtered_df[target_col].value_counts().reset_index()
        value_counts_df.columns = [f'{target_col}', 'Value Counts']
        print(f'Value Counts for {target_col} (Hue {hue_col} = {hue_value}):\n')
        print(value_counts_df)
        print('='*40)

    plt.suptitle(f'Influence of {hue_col} on {target_col}', fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust layout to accommodate titles
    plt.show();
```

```python
#create function to plot evolution of claim numbers cat (or other categorical columns) over time period of other column
def plot_timeline_cat(df=contracts, y_col='num_claims_cat', date_col='policy_startDate'):
    #get dummies of y_col for grouping
    df = pd.get_dummies(contracts[y_col])
    
    #create dict for agg columns while grouping
    map = {}
    for col in df.columns:
        map[col] = 'sum'
    
    #create year column out of date col to groupby
    df['year'] = contracts[date_col].dt.year
    
    grouped_df = df.groupby('year').agg(map)
    sns.lineplot(data=grouped_df)
    plt.ylabel(f'{y_col} per category')
    plt.title(f'Evolution of {y_col} over the years of {date_col}');
    print(grouped_df)
```
